```python
# functions that can compare 2 numbers
def compare(x,y):
    """
    This is a function to compare 2 numbers, x and y
    """
    # if statement
    if x > y:
        print(f"{x} is greater than {y}")
    elif x < y:
        print(f"{x} is smaller than {y}")
    else:
        print(f"{x} is the same as {y}")
```

```python
compare(9, 0)
```

```python
compare(10, 0)
```

## Create custom images

```python
import matplotlib.pyplot as plt
```

```python
#create a balck grayscale image
black = np.zeros((200, 100), dtype=np.uint8) #"uint8"

plt.imshow(black, cmap=plt.cm.gray) #"gray"
plt.title("black image")
plt.xticks([]), plt.yticks([])
plt.show
```

```python
# Create a white grayscale image
white = np.zeros((200, 100)) + 255
white = np.uint8(white)
# white = white.astype("uint8")

plt.imshow(white, cmap=plt.cm.gray, vmin=0, vmax=256) #"gray"
plt.title("white image")
plt.xticks([]), plt.yticks([])
plt.show
```

```python
#random color image
random_color_img = np.random.randint(0, high=256, size=(150, 250, 3), dtype = np.uint8)

plt.imshow(random_color_img) #"gray"
plt.title("random color image")
plt.xticks([]), plt.yticks([])
plt.show
```

```python
def OddEven(x):
    """This function is to evaluate the odd even nature of number"""
    if isinstance(x, int):
        if x%2 == 0:
            if x == 0: 
                statement = "is neither odd or even number." 
            else: 
                statement = "is even number."
        else:
            statement = "is odd number." 
    else:
        statement = "is not integer." 
    return print(f"{x}" + statement)
```

```python
OddEven(0)
```

```python
OddEven(5.7)
```

```python
OddEven(201)
```

## Exercise 4

```python
import numpy as np
import matplotlib.pyplot as plt

image_size = 200
image = np.zeros((image_size, image_size), dtype=np.uint8)

center_x, center_y = image_size // 2, image_size // 2
radius = 50

for x in range(image_size):
    for y in range(image_size):
        if (x - center_x) ** 2 + (y - center_y) ** 2 < radius ** 2:
            image[y, x] = 255

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```
